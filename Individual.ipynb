## PART 3: EXPLORATORY DATA ANALYSIS AND VISUALIZATION ##

The quantitative values in this data set are Age and played_hours, both in the double or "dbl" form. To compute the mean for both age and hours played, I can use the `mean()` function with `summarize()`:

```r
players_mean <- players1 |>
  summarise(mean_age = mean(Age, na.rm = TRUE), mean_time_played_hrs = mean(played_hours, na.rm = TRUE))

players_mean
```

This shows that the mean age of players is 20.52 years and the mean hours played is 5.85 hours.

Now, creating a visualization to try and see if there might be a relationship between age, gender, and hours played on the MineCraft server. Firstly, I will make a bar graph to see the total hours played per age. 

```r
players <- players1 |>
    select(Age, gender, played_hours)
```

```r
head(players) #we can now see the first 6 rows of the new selected data set that I will use to address my question
```

```r
options(repr.plot.width = 8, repr.plot.height = 7)
age_plot <- players|>
    ggplot(aes(x = Age, y = played_hours, fill = gender))+
            geom_bar(stat = "identity") +
    labs(x = "Age (in years)", y = "Hours played", title = "Graph 1: Age vs. Hours Played Bar Graph", fill = "Gender") +
    theme(element_text(size = 15))
age_plot
```

From this bar graph, we can see that the ages betwen 16-20 have the highest number of hours played collectively. Using the legend, we can also see that there is a large proportion of Male and Female players. Let's explore this further in another bar graph that focuses on gender. 

```r
options(repr.plot.width = 8, repr.plot.height = 7)
gender_plot <- players|>
    ggplot(aes(x = gender, y = played_hours, fill = gender))+
            geom_bar(stat = "identity") +
    labs(x = "Gender", y = "Hours played", title = "Graph2: Gender vs. Hours Played Bar Graph", fill = "Gender")+
    theme(element_text(size = 15))
gender_plot
```

Now we can see more clearly that there are many Male and Female players. This will be very useful in addressing the  question.

```r
options(repr.plot.width = 8, repr.plot.height = 7)
age_scatterplot <- players|>
    ggplot(aes(x = Age, y = played_hours, colour = gender))+
            geom_point() +
    labs(x = "Age (in years)", y = "Hours played", title = "Graph 3: Age vs. Hours Played Bar Graph", colour = "Gender") +
    theme(element_text(size = 15))
age_scatterplot
```

We can see from this scatter plot that the trend between age and hours played is not linear. This will be useful in determining the right prediction model. 

## PART 4: METHODS AND PLAN ##

**Why is this method appropriate?**

To address the question "can age and gender predict the amount of hours played?", a knn-regression model (K nearest neighbours regression) will be used. This was chosen because the trend does not appear to be linear, as shown in Graph 3, and that hours played is quantitative, so we need to predict a quantitative variable. Thus, knn-regression will do the best job at predicting the hours played based on the predictors age and gender. 

**Which assumptions are required, if any, to apply the method selected?**

We will need to assume that the data is reported correctly by each participant and that the data is standardized. Just in case, the data will be standardized before any analysis is done when addressing the question.

**What are the potential limitations or weaknesses of the method selected?**
Knn-regression can be slow as training data gets larger, so I will need to make sure that the training set is sufficiently large to provide an accurate trained model, but not overly large so that the regression is too slow. It also does not perform well with many predictors, but since we are only covering age and gender, that should not be an issue. Lastly, the model is less reliable beyond the range of values inputted from the data. This means that extrapolation should not be done, since it will likely not provide an accurate prediction. 

**How are you going to compare and select the model?**
A cross validation will be used, which will train the model to find the best K to predict the hours played. I will also find the RMSE (root mean squared error) to assess how well the model fits the training data before using it on the testing data. Then, I will find the RMSPE (root mean squared predicted error) to see if the model also fits the testing data well.

**How are you going to process the data to apply the model?**
To process the data, it will first have the NAs removed from the Age column. Then, the data will initially be split into training and testing sets (75% and 25% respectively). After, a cross-validation using 5 folds will take place to find the best K value for the predictive model using the training set. Once the best K has been determined, it will be used on the testing set and its RMSE will be found to see how well it fits the training set. Then, the model will be used for the testing set and its RMSPE will be determined to see how well the model fits its data.
